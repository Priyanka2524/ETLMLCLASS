Step 1 - Data Ingestion 

```python
import pandas as pd
df = pd.read_csv("drug200.csv")
df.head()
```

Target Feature - Drug

Step 2 - Perform basic data quality checks

```python
df.info()
```

Step 3 - Seperate X and Y 

```python
X = df.drop(columns=["Drug"])
Y = df[["Drug"]]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
Y.value_counts().plot(kind="bar")
```

Step 4 - Apply preprocessing on X

```python
X.dtypes
```

```python
cat = list(X.columns[X.dtypes == "object"])
con = list (X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore",sparse_output=False)
)
```

```python
pre = ColumnTransformer(
    [
        ("num", num_pipe,con),
        ("cat",cat_pipe,cat)
    ]
).set_output(transform="pandas")
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

Divide data in train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain,ytest = train_test_split(X_pre, Y, test_size=0.4,random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

Step 6 - Build the decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(
    max_depth=1,
    min_samples_split=5,
    min_samples_leaf=5,
    criterion="entropy"
)
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

# Hyperparameter tuning

```python
params ={
    "max_depth": [2,3,4,5,6],
    "min_samples_split":[5,6,7,8,9],
    "criterion" : ["gini","entopy"]
}
```

```python
 from sklearn.model_selection import GridSearchCV
```

```python
base_model = DecisionTreeClassifier()
gscv = GridSearchCV(base_model,param_grid=params, cv=5,scoring="f1_macro")

gscv.fit(xtrain,ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_dtc = gscv.best_estimator_
best_dtc
```

# Step 7 - Model Evaluation 

```python
best_dtc.score(xtrain, ytrain)
```

```python
best_dtc.score(xtest,ytest)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(best_dtc,xtest, ytest)
```

```python
ypred_test = best_dtc.predict(xtest)
ypred_test[0:5]
```

```python
ytest.head()
```

```python
yprob_test = best_dtc.predict_proba(xtest)
yprob_test[0:5]
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred_test))
```
